```python
students_names=["sanjay","Rahul","Karan","Wasim","Ramesh","Ajay","Sartaj","Priya"]

students_marks=[35,50,20,45,25,40,25,40]
```

```python
marks_perc = []
for x in students_marks:
    res = (x/50)*100
    marks_perc.append(res)
marks_perc
```

```python
def line_chart_of_studends_and_mark():
    plt.plot( students_names,students_marks, marker= "o", mec = "k", ms=15, linestyle= "dashed", color="purple", linewidth=3)
    
    plt.title('Students Mark Graph')
    plt.xlabel('Students Name')
    plt.ylabel('Students Marks')
    plt.show()
    
line_chart_of_studends_and_mark()
```

```python
def percentage_bar_chart():
    plt.bar( students_names,marks_perc,color="purple")
    
    plt.title('Students Percentage Graph')
    plt.xlabel('Students Name')
    plt.ylabel('Students Percentage')
    plt.show()
    
percentage_bar_chart()
```
